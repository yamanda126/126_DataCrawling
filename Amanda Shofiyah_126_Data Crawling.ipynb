```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import time
```

```python
# INSTALL BROWSER SUCH CHROME OR EDGE\
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER SUCH FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []
default_value = "None"  # Nilai default untuk kolom yang kosong

for index in range(1, 40):
    browser.get(f'https://www.zalora.co.id/c/luxury/tas-wanita-dompet-luxury/c-11005?csa=topcategoriesbags&sort=popularity&page={index}')
    time.sleep(30)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='flex flex-col')
    
    for row in rows:
        merk_element = row.find(class_='flex flex-1 items-center truncate')
        merk = merk_element.text if merk_element is not None else default_value
    
        cashback_element = row.find(class_='truncate text-xxs font-bold tablet:text-xs')
        cashback = cashback_element.text if cashback_element is not None else default_value

        min_pembelian_element = row.find(class_='truncate text-xxxs leading-3 tablet:text-xxs')
        min_pembelian = min_pembelian_element.text if min_pembelian_element is not None else default_value

        tipe_element = row.find(class_='overflow-hidden text-ellipsis whitespace-nowrap text-sm')
        tipe = tipe_element.text if tipe_element is not None else default_value

        disc_element = row.find(class_='text-sm leading-6 text-grey-100')
        disc = disc_element.text if disc_element is not None else default_value

        before_disc_element = row.find(class_='text-grey-60 line-through')
        before_disc = before_disc_element.text if before_disc_element is not None else default_value

        after_disc_element = row.find(class_='font-bold text-red-60')
        after_disc = after_disc_element.text if after_disc_element is not None else default_value

        harga_Fix_element = row.find(class_='zui-tag zui-tag__alert inline-flex items-center whitespace-nowrap rounded-lg border-[0.5px] border-red-60 bg-white h-[18px] gap-1 px-1')
        harga_Fix = harga_Fix_element.text if harga_Fix_element is not None else default_value
        if harga_Fix == default_value:
            harga_Fix = after_disc
        elif after_disc != default_value:
            harga_Fix += " + " + after_disc

        dt = {
            'merk': merk,
            'cashback': cashback,
            'min_pembelian': min_pembelian,
            'tipe': tipe,
            'disc': disc,
            'before_disc': before_disc,
            'after_disc': after_disc,
            'harga_Fix': harga_Fix
            }
        data.append(dt)
```

```python
df = pd.DataFrame(data)
```
